```python
# importing pandas

import pandas as pd
```

```python
# Read salaries.csv in pandas

sal = pd.read_csv('salaries.csv')
```

```python
# Print the head of data frame sal

sal.head()
```

```python
#  Use .info method to find out how many entries are there

sal.info()
```

```python
# what is the average base pay

sal.BasePay.mean()
```

```python
# What is the Highest amount of overtimePay in the dataset

sal.OvertimePay.max()
```

```python
# What is the Job Title of JOSEPH DRISCOLL ? 

sal[sal['EmployeeName'] == "JOSEPH DRISCOLL"]["JobTitle"]
```

```python
#  How much does Joseph Driscoll make inlcuding benifits

print(sal[sal['EmployeeName']=="JOSEPH DRISCOLL"]['TotalPay'])
```

```python
# What is the name of highest paid person including benifits

# sal[sal['TotalPay'].max()]['EmployeeName']
sal[sal['TotalPayBenefits']==sal.TotalPayBenefits.max()]['EmployeeName']
```

```python
# What is the name of lowest paid person inlcuding benifits
sal[sal['TotalPayBenefits']==sal.TotalPayBenefits.min()]['EmployeeName']
```

```python
# What is the avg (mean) base pay of all employees per year 2011-2014
sal.groupby('Year').mean().BasePay
```

```python
# How many Unique Job Titles are there 

sal['JobTitle'].nunique()
```

```python
# What are the top 5 most common jobs ?
sal['JobTitle'].value_counts().head()
```

```python
#  How many Job Titles were Respresented by only 1 person in 2013 ?

sum(sal[sal['Year']==2013]["JobTitle"].value_counts()==1)
```

```python
# How many people have chief in thier job title?

def chief_string(title):
    if "chief" in title.lower().split():
        return True
    else:
        return False
```
